```python
# SOLUTION:  Cleaning the data

##First we will define a function to clean the Age column
##All ages should be integers
def clean_age(ageString):
    # For safety, if the passed value is missing, don't change it 
    if pd.isnull(ageString):
        return ageString 
    
    # With strings, it's a good idea to call strip() to remove
    # leading or trailing white space 
    stripped_val = ageString.strip()
    
    # Use string.replace( ) to replace the non numerical characters 
    # with blanks ('') 
    num_str = stripped_val.replace('A', '')
    
    # After replacing the errant characters, we still have a string 
    # We need to convert it to a floating point value 
    ageNum = int(num_str)
    
    return ageNum

#now call the apply() function on the Age column 
dfFrosh.loc[:, "age"] = dfFrosh.loc[:,"age"].apply(clean_age)



##Now I will define a function to clean the lunch money column
#Should be rational numbers measured to the hundredths place to represent dollars and cents.
def clean_money(moneyString):
    # For safety, if the passed value is missing, don't change it 
    if pd.isnull(moneyString):
        return moneyString 
    
    # With strings, it's a good idea to call strip() to remove
    # leading or trailing white space 
    stripped_val = moneyString.strip()
    
    # Use string.replace( ) to replace the non numerical characters 
    # with blanks ('') 
    num_str = stripped_val.replace('$', '')
    
    # After replacing the errant characters, we still have a string 
    # We need to convert it to a floating point value 
    floatMoney = float(num_str)
    
    return floatMoney

#now call the apply() function on the lunch money column 
dfFrosh.loc[:, "lunch_money"] = dfFrosh.loc[:,"lunch_money"].apply(clean_money)


##lastly I will round the height column and the lunch money column
#heights should be measured to the tenths place
#lunch money should be measured to the hundreths place
#so I will use the round function on the data frame and round the height column to 1 decimal point 
#and the lunch money column to 2 decimal points
dfFrosh = dfFrosh.round({"height":1, "lunch_money":2})
```

- If you grab a Coke first, then you will drop that disgusting drink back into the icy water of the cooler and add 2 more warm clam juices to the cooler in order to increase the ratio for the next grab.

## Part A



***(10 points)*** Write a simple simulation with 10,000 trials that estimates the probability that your ***first and second grabs are both Cokes***.

```python
drinks = np.array(["coke", "clam"])
# 'drinks' is an array holding the two strings "coke", "clam".
num_samples = 10000 #run 10,000 trials

# pick the first drink 
p1 = np.random.choice(drinks, size=num_samples)
# so 'p1' is an array that holds num_samples amount of
#  "coke" or "clam" from the 'drinks' array.

# do the sampling as a .apply() to df of first pick
df = pd.DataFrame({"Pick 1" : p1})
  # df contains num_samples random selections of  "coke", "clam".
    
    
# this function picks a second drink based on result of the first one
# takes 'row' of df and 'some_set'  'drinks' which hold our two drink options
def pick_second(row, some_set):
    return np.random.choice(some_set, p=[12/26,14/26]) if row[0]=="coke" else np.random.choice(some_set, p=[1/2, 1/2])

# actually apply the function
# axis=1 is the 'row' the function is taking as input
df["Pick 2"] = df.apply(pick_second, axis=1, some_set=drinks)

# compute ratio of both cokes out of total samples
both_coke_prob = len(df.loc[(df["Pick 1"]=="coke") & (df["Pick 2"]=="coke")])/num_samples
print("The probability that both picks are coke is approximately {:.3f}".format(both_coke_prob))
```

**By hand verification**: Let $K_1$ be the event that a coke is grabbed first and $K_2$ be the event that a coke is grabbed second.  We want to compute $P(K_1 \cap K_2)$.  We'll do this using the product rule 

$$
P(K_1 \cap K_2) = P(K_2 \mid K_1)~P(K_1) 
$$

Now, clearly the probability of  a coke being grabbed first is $P(K_1) = \frac{1}{2}$.  The conditional probability that a coke is grbbed second _given_ that a coke was grabbed first is $P(K_2 \mid K_1) = \frac{12}{26}$. 

Then the probability of both picks being coke 

$$
P(K_1 \cap K_2) = P(K_2 \mid K_1) ~ P(K_1) = \frac{12}{26}\cdot\frac{1}{2} = \frac{3}{13} = 0.2307
$$

which seems to agree very well with our simulation!

```python
##function to calculate the PMF for a binomial distribution
def binomial_pmf(n, p): 
    pmf = np.array([binom(n,k)*(p**k)*(1-p)**(n-k) for k in range(0,n+1)])
    return pmf

##function to calculate the CDF for a binomial distribution
def cdf_clamjuice(n,p):
    pmf = binomial_pmf(n,p) #using the above function to get the pmf 
    cdf=np.cumsum(pmf) #cumulative sum of pmf to get cdf
    
    ##print the results
    print("CDF: The probability of Leroy getting a clam juice...")
    for k, prob in enumerate(cdf):
        print("{} times or fewer is {:0.4f}".format(k,prob))
    
    #return the np.array
    return cdf 
```

```python
cdf= cdf_clamjuice(10,.75)

#graph representation (I know is not required but nice to visualize)
x = np.array([0,1,2,4,5,6,7,8,9,10,11])
y = np.array(cdf)
plt.stem(x, y)
plt.xlabel("X times or fewer")
plt.ylabel("Probability")
plt.title("CDF of Leroy getting clam juice")
plt.show()
```
