### Exercise 3 - Polynomial Regression via Multiple Linear Regression 
*** 

It's not too difficult to believe that some relationships between features and the response are $\color{red}{\text{nonlinear}}$.

Consider the following example, where the single feature $x$ and the response $y$ have a quadratic relationship of the form 

$$
Y = \frac{1}{4} - X + X^2 + \epsilon 
$$

The following code samples $n=30$ data points from the true model,

fits the SLR model (i.e. a line),

and plots it against the data.

It should be clear from both the picture and the $\color{red}{R^2}$-value that the SLR model is a very poor fit of the data. 

```python
# Generate Data

# There will be n = 30 data points
n = 30

# Standard Deviation
sig = 0.05

# 'x' is an array containing values from a N(linspace, 0.05^2)
x = np.linspace(0,1,n) + sig*np.random.rand(n)
# Example: 3 + 2.5 * np.random.randn() is an array of samples from N(3, 6.25).
#x = np.linspace(0,1,n)  # This is an optional array of x-values.

# 'y' is an array containing noisy values about our parabola.
y = 0.25 - x + x**2 + stats.norm.rvs(0,sig,size=n)

# Fit SLR model, i.e. a line, to our parabolic data cloud. 
bhat, ahat, rval, pval, stderr = stats.linregress(x, y)

# Now, plot the data and the SLR model (line).

# Initialize the plot.
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,6))

# 'xplot' is an array holding evenly spaced values between the min and max of 'x'.
xplot = np.linspace(np.min(x), np.max(x)) # 50 values by default.

# Plot the regression line, using 'xplot'.
ax.plot(xplot, ahat + bhat*xplot, color="steelblue", lw=3)

# Scatter plot of 'x' and 'y'; the parabolic cloud.
ax.scatter(x, y, color="steelblue", s=100, alpha=0.9)
# s = 100 is for marker size. alpha =0.9 is for transparency color of marker.

# Make the graph pretty.
ax.grid(alpha=0.25)
ax.set_xlabel("X", fontsize=16)
ax.set_ylabel("Y", fontsize=16)
ax.text(0.8, np.min(y)+.00, r"$R^2$={:.5f}".format(rval**2), fontsize=20);
```

**Part A**: We can fit a **polynomial** model to the single-feature data by thinking of the polynomial features as features in a multiple linear regression model.  If we make the association  

$$
\color{red}{x_1} = x \quad \textrm{and} \quad \color{red}{x_2} = x^2 
$$

Imagine: $\phantom{x}\color{red}{x_1}$: 1, 2, 3, 4, ...

$\phantom{xxxxxxxxx}\color{red}{x_2}$: 1, 4, 9, 16, ... 

then we can fit a multiple $\color{blue}{\text{linear regression}}$ of the form 

$$
\hat{y} = \hat{\beta}_0 + \hat{\beta}_1 x_1 + \hat{\beta}_2 x_2 
$$

Let's see how we can do this in python. 

```python
# Using 'np.column_stack'

# a = np.array((1,2,3))
# b = np.array((2,3,4))

# np.column_stack((a,b))

# array([[1, 2],
#        [2, 3],
#        [3, 4]])

# Collect the linear feature x and it's square in an array 
X = np.column_stack((x, x**2))

# Add a constant to the array for the intecept 
X = sm.add_constant(X)

# Fit the ordinary least-squares (OLS) model 
polymodel = sm.OLS(y, X).fit()

# Print estimated parameters 
print(polymodel.params)
```

**Part B**: Write down the estimated MLR model in terms of the features $x_1$ and $x_2$ as well as the interpretation of the associated polynomial model in terms of the single feature $x$. Does this model seem close to the true model that the data was generated from?  

**Solution**: We have 

$$
\hat{y} = \color{blue}{0.2503} \color{red}{- 0.9347} x_1 + \color{red}{0.9507} x_2 \quad \textrm{and} \quad 
\hat{y} = \color{blue}{0.2503} \color{red}{- 0.9347} x + \color{red}{0.9507} x^2 
$$

The true model was $y=\color{blue}{0.25} \color{red}{- 1} x + \color{red}{1} x^2$.

So, yes, it seems very close to the true model that the data was generated from. 

**Part C**: Modify the code below to $\color{blue}{\text{plot}}$ the obtained $\color{blue}{\text{polynomial regression model}}$ against the data.  How does it look?  

```python
# Plot data and "SLR" model 

# Initialize the plot.
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,6))

# 'xplot' contains x-values evenly spaced between the min and the max of 'x',
#  where x = np.linspace(0,1,n) + sig*np.random.rand(n)
xplot = np.linspace(np.min(x), np.max(x))

# 'yplot' is an array of y-values that live on the MLR model we generated.
yplot = polymodel.params[0] + polymodel.params[1]*xplot + polymodel.params[2]*xplot**2

# Plot the MLR paraboa.
ax.plot(xplot, yplot, color="steelblue", lw=3)

# Create a scatter plot of the 'x' and 'y' parabolic cloud.
ax.scatter(x, y, color="steelblue", s=100, alpha=0.9)

# Make the graph pretty.
ax.grid(alpha=0.25)
ax.set_xlabel("X", fontsize=16)
ax.set_ylabel("Y", fontsize=16);
```

It looks pretty good! 

**Part D:** Calculate the sum of squared errors (SSE),

the total sum of squares (SST),

and the coefficient of determination ($R^2$) for this polynomial regression model.

How does the $R^2$ value compare to the linear regression of **Part A**?

```python
# Note 'polymodel = sm.OLS(y, X).fit()' is our parabolic model, therefore,
#  polymodel.params provides our model parameters.
# 'params' is an array holding our 3 parameters with indices 0, 1, 2.


# This function takes an array of x-values and returns an array of the
#  parabola of y-values from using our MLR model parameters.
def quadfit(x, params):
    return params[0] + params[1]*x + params[2]*x**2

# 'yhat' contains the parabola of y-values by calling 'quadfit'.
yhat = quadfit(x, polymodel.params)

# 'ybar' contains the mean of the y-values from the parabolic cloud of y-values,
#   generated with y = 0.25 - x + x**2 + stats.norm.rvs(0,sig,size=n).
ybar = np.mean(y)

# sum of squared errors; for calculating R^2
sse = np.sum((y-yhat)**2)

# total sum of squares; for calculating R^2.
sst = np.sum((y-ybar)**2)

# Our hand calculated coefficient of determination.
R2 = 1 - sse/sst

print('SSE = {:0.3f}, SST = {:.3f}, and R^2 = {:.3f}'.format(sse,sst,R2))
```

Solution:

The $R^2$ value from part A was near 0 because we were attempting to fit a line to the parabolic shaped data. This $R^2$ value comes from attempting to fit a parabola to the parabolic shape and is of course a much higher value.

In the previous exercise we looked for a polynomial regression model that was quadratic because we already knew the true form of the model (which never happens in practice).

Recall that the residual is the difference between the true response value $y$ and the response predicted by the model, $\hat{y}$: 

$$
r_i = y_i - \hat{y}_i 
$$

### Helper Functions
***

```python
def nonlinear_data(dataset, n=30):
    
    if dataset < 0 or dataset > 2: 
        print("Datasets must be numbered 0-2.  Defaulting to 0.")
        dataset = 0
    
    np.random.seed(1237)
    
    X = np.linspace(0,1,n) + .05 * np.random.rand(n)
    
    if dataset == 0: 
        return X, 0.5 + 0.75 * X + .5*np.random.rand(n)
    elif dataset == 1: 
        return X, 0.25 - X + X*X + .1*np.random.rand(n)
    elif dataset == 2: 
        return X, 2*(3*(2*x-1.2)**3 + 2*(2*x-1.2)**2 -(2*x-1.2)) + 1.5*np.random.rand(n)

def fit_and_res_plot(X, y, hint=False):
    
    if len(X.shape)==1:
        x = X
    else:
        x = X[:,0]
    
    # Fit Data 
    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit() 
    
    # Print fitted model 
    print("Fitted PLR Model: yhat = ", end="")
    for p, beta in enumerate(model.params):
        print("{:.3f}".format(beta), end="")
        if p > 0: 
            print(" x^{} ".format(p), end="")
        else:
            print(" ", end="")
        if p != len(model.params)-1:
            print("+ ", end="")
            
    # Plot data/fit 
    yhat = model.predict(X)
    res = y - yhat 
    xp = np.linspace(np.min(x), np.max(x), 50)
    yp = np.zeros(len(xp))
    for p, beta in enumerate(model.params):
        yp += beta * xp**p 
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16,6))
    axes[0].scatter(x, y, color="steelblue", s=100, alpha=0.9)
    axes[0].plot(xp, yp, color="steelblue", lw=3)
    axes[0].grid(alpha=0.25)
    axes[0].set_xlabel("X", fontsize=16)
    axes[0].set_ylabel("Y", fontsize=16)
    axes[0].set_title("Data and PLR Fit", fontsize=16)
    
    # Plot residuals 
    axes[1].plot([np.min(x), np.max(x)], [0, 0], lw=2, ls="--", alpha=0.5, color="black")
    axes[1].scatter(x, res, color="#6a9373", s=100, alpha=0.9)
    axes[1].grid(alpha=0.25)
    axes[1].set_xlabel("X", fontsize=16)
    axes[1].set_ylabel("Res", fontsize=16)
    axes[1].set_title("Residual Plot", fontsize=16)
    
```
