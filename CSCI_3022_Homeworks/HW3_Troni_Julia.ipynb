$$ P(B \cup C) = P(B) + P(C) - P(B \cap C) $$
$$ P(A \cup C) = P(A) + P(C) - P(A \cap C) $$

Events A and C, and events B and C are independent so 
$$ P(B \cap C)= P(B)\cdot P(C)$$ and $$P(A \cap C)= P(A)\cdot P(C)$$


So, $$ P(B \cup C) = P(B) + P(C) - P(B)\cdot P(C) $$

So $$ P(A \cup B \cup C)= P(A) + P(B \cup C)- P(A) \cdot P(C) $$
We can reduce to
$$ \frac{11}{12} = P(A) + \frac{9}{12} -P(A) \cdot P(C) $$ 
$$ \frac{2}{12} = P(A)-P(A) \cdot P(C) $$

# Problem 3

It is free donut day on campus! There are three locations offering each participant one free donut; either a glazed donut or a cake donut.

The C4C is offering donuts, 75% of their donuts are glazed and 25% are cake.

The Alfred Packer Grill is offering donuts. 60% of their donuts are glazed and 40% are cake.

The SEEC Cafe (East campus) is offering donuts, 45% are glazed and 55% are cake.

# Part A

***(4 points)***

If you randomly choose a donut location (each location is equally likely) and then your chosen location randomly grabs a donut and hands it to you, then what is the probability that your newly acquired donut is glazed?

$\color{blue}{\text{The picture below is merely here to help visualize this as a law of total probability type problem.}}$

$\color{blue}{\text{You need not use the pictured variables.}}$

![title](tp.png)

***solution:***
$$ P(Glazed)= \frac{P(Glazed Donut)}{P(Total Donuts)} $$

$$ P(Glazed)= \frac{P(Glazed C4C)+P(Glazed Alfred Packer)+P(Glazed SEEC)}{P(Total Donuts)}= \frac{75+60+45}{300} = \frac{180}{300} = 0.6 $$

# Part B

***(5 points)***

For the same situation mentioned above, create some code that will run a simulation to estimate the probability of getting a glazed donut. *(Use n=1000)*

Run the code and verify that it agrees with the by-hand computation you arrived at above.

```python
C4C = {'donuts' : np.array(["glazed", "cake"]), 'probs' : np.array([75/100, 25/100])}
AlfredPacker = {'donuts' : np.array(["glazed", "cake"]), 'probs' : np.array([60/100, 40/100])}
SEEC = {'donuts' : np.array(["glazed", "cake"]), 'probs' : np.array([45/100, 55/100])}

loc_choices = {'locations' : np.array([C4C, AlfredPacker, SEEC]), 'probs' : np.array([1/3, 1/3, 1/3])}

def sample_donut(loc_choices):
    # randomly choose a donut location
    location = np.random.choice(loc_choices['locations'], p = loc_choices['probs'])
    # randomly choose a donut from that location
    return np.random.choice(location['donuts'], p = location['probs'])

def probability_of_donut(variety, loc_choices, num_samples=1000):
    # get a bunch of donuts
    donuts = np.array([sample_donut(loc_choices) for ii in range(num_samples)])
    # compute fraction of donuts of desired type
    return np.sum(donuts == variety) / num_samples
```

```python
probability_of_donut("glazed", loc_choices, num_samples=50000)
```

The code and the by-hand computation agree. Both show the probability that your newly acquired donut is glazed is 60%

# Problem 5
***
In an attempt to avoid COVID, a group of 100 students seperate themselves from CU society and choose to live in connected housing apart from all other people.

Unfortunately, and unbeknown to anyone, one of the students had COVID when the group went into isolation.

It is known that 90% of COVID tests will detect the virus (i.e. true positives) while 9.6% of the tests are false positives.

***(7 points)***

If, on move-in day, a randomly chosen person, within this group of 100, gets a positive test result, what is the probability that they actually have COVID?

***solution:***
Let the events:
- V= virus is actually in the person
- T= test is positive (test says the person has the virus) 

Then $$P(true positive)= P(T|V)= \frac{90}{100}$$
$$P(false positive)= P(T|V^C)= \frac{96}{1000}$$

Then since $$P(T|V)+ P(T^C|V)= 1$$ and $$P(T|V^C)+ P(T^C|V^C)= 1$$
$$P(false negative)= P(T^C|V)= \frac{10}{100}$$
$$P(true negative)= P(T^C|V^C)= \frac{904}{1000}$$

We know that $$P(V)= \frac{1}{100}$$ so $$P(V^C)= \frac{99}{100}$$
We are being asked to find $$P(V|T)$$

$$P(V|T)= \frac{P(V \cap T)}{P(T)} = \frac{P(T|V)\cdot P(V)}{P(T|V)\cdot P(V)+ P(T|V^C)\cdot P(V^C)} $$
$$P(V|T)= \frac{\frac{90}{100} \cdot \frac{1}{100} }{ \frac{90}{100} \cdot \frac{1}{100} + \frac{96}{1000} \cdot \frac{99}{100} } $$
$$P(V|T)= 0.0865051903114187 $$

Therefore there is a 8.65% chance that the randomly chosen person with the positive test result actually has COVID.
